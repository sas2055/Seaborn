# Seaborn 
    - Seaborn is an open source, BSD-licensed Python library providing high level API for visualizing the data using Python programming language.
    -It is built on top of Matplotlib.

### Seaborn Vs Matplotlib
    - It is summarized that if Matplotlib “tries to make easy things easy and hard things possible”, Seaborn tries to make a well-defined set of hard things easy too.”

    - Seaborn helps resolve the two major problems faced by Matplotlib; the problems are −

    1. Default Matplotlib parameters
    2. Working with data frames

## Plotting using Seaborn

#### Histogram using distance plotting:
    - Histograms represent the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

```python
df
```

```python
sns.histplot(x='sepal_length',data=df)
```

```python
sns.histplot(x='sepal_length',data=df,hue='species')
```

```python
# supply own ranges using bins
sns.histplot(x='sepal_length',data=df,bins=[5.0,6.0,7.0])
```

```python
# supply own ranges using bins
sns.histplot(x='sepal_length',data=df,bins=4)
```

#### Flexibly plot a univariate distribution of observations.

#### displot

```python
sns.displot(x='sepal_width',data=df)
```

```python
sns.displot(df.petal_length)
```

#### distplot:
    - distplot goves u 2 plot in a combination
    1. hist
    2. KDE

```python
sns.distplot(df.petal_length)
```

#### now only i want hist plot so will make kde=False

```python
sns.distplot(df.petal_length,kde=False)
```

```python
# if i want all columns to be involved in plotting
sns.displot(data=df)
```

```python
ndf = df[['sepal_length','sepal_width']]
sns.displot(data=ndf,palette='CMRmap_r')
```

#### KDE:
    - Kernel Density Estimation (KDE) is a way to estimate the probability density function of a continuous random variable. 
    - It is used for non-parametric analysis.
    - If we want combinition of hist and kde then ...
    - Univariant distribustion

```python
sns.kdeplot(df.petal_length)
```

```python
sns.kdeplot(df.petal_width)
```

```python
sns.kdeplot(data=df)
# assigning a label using matplotlib
plt.xlabel('Input data')
plt.ylabel('Density of Data')
plt.title('Sample graph')
```

```python
a = sns.kdeplot(data=df)
# assigning a label using seaborn
a.set_xlabel('Input data')
a.set_title('Sample Graph')
```

```python
a = sns.kdeplot(data=ndf)
# assigning a label using seaborn
a.set_xlabel('Input data')
a.set_title('Sample Graph')
```

```python
b = sns.kdeplot(data=pdf)
# assigning a label using seaborn
b.set_xlabel('Input data')
b.set_title('Sample Graph')
```

## Univariate bivariate multivariate??

#### Plotting Bivariate Distribution: Joint Plot
    - Bivariate Distribution is used to determine the relation between two variables.

```python
df[:2]
```

```python
# It gives scatter plot & its strictly bivariate plot
# + it give hist plot on axes spines
sns.jointplot(x='sepal_length',y='sepal_width', data=df)
```

```python
# use kind= hex
sns.jointplot(x='sepal_length',y='sepal_width', data=df,kind='hex')
```

```python
# use regression 
sns.jointplot(x='sepal_length',y='sepal_width', data=df,kind='reg')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width', data=df,hue='species')
```

```python
sns.jointplot(x='petal_length',y='sepal_width', data=df,hue='species')
```
