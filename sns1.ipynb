```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.head(2)
```

#### Kernel Density Estimation:
    - Kernel density estimation is a non-parametric way to estimate the distribution of a variable.

```python
sns.kdeplot(df.petal_length)
```

```python
sns.kdeplot(x='sepal_width',data=df)
```

```python
dfn = df.drop(columns=['species'])
dfn[:2]
```

```python
sns.displot(df)
```

```python
sns.kdeplot(data=dfn)
```

#### Pair relationships using pairplot:
    - To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function.

```python
# pairplot works on Numeric data only
sns.pairplot(df)
```

```python
sns.pairplot(df,kind='reg')
```

```python
# hue is used on categorical data
sns.pairplot(df,hue='species')
```
